2. 문자열에서 숫자 추출 후 해당값 정수형으로 변환하기 (함수까지 만들기까지)

```python
# 텍스트 안에서 숫자만 추출
    # 반복문으로 한글자씩 출력
    # 한글자씩 조건문으로 확인
    # 조건문에 해당하면, 해당 글자를 정수형으로 변환
# 문자열 타입인 숫자를 정수형으로 변환


테스트 = "3일 전"

# for 숫자데이터 in 테스트:
#     if 숫자데이터.isdigit():
#         # print(숫자데이터)
#         변환_숫자데이터 = int(숫자데이터)
# 변환_숫자데이터

# type(변환_숫자데이터)

# 반복해서 위 기능을 사용하고 싶어.
# 어떤 관점을 갖고 코드를 봐야 재사용 할 수 있게 만들수 있을까?
# --> 테스트 한 대상을 파라미터로 생각하고 코드를 수정한다. 
      #그리고 출력하고 싶은 것 왼쪽에 return 붙여라

def 숫자변환(날짜넣으시오):
    for 숫자데이터 in 날짜넣으시오:
        if 숫자데이터.isdigit():
            # print(숫자데이터)
            변환_숫자데이터 = int(숫자데이터)
    return 변환_숫자데이터

숫자변환(테스트)
```

3. 변환한 정수형 숫자를 현재날짜에서 빼기 (함수까지 만들기)

```python
# 문자열에서 숫자 추출 후 정수형으로 변환한 값을 날짜 형식으로 바꾸기
    # 1. 숫자 추출 후 정수형으로 변환
    # 2. 현재 날짜에서 정수형으로 변환한 거 빼기
    # 3. 뺀 값을 보기 좋게 변환

테스트 = "3일 전"

# 1. 숫자 추출 후 정수형으로 변환
for 숫자데이터 in 테스트:
    if 숫자데이터.isdigit():
        # print(숫자데이터)
        변환_숫자데이터 = int(숫자데이터)

# 2. 현재 날짜에서 정수형으로 변환한 거 빼기
날짜로변환 = 현재날짜 - timedelta(days=변환_숫자데이터)

# 3. 뺀 값을 보기 좋게 변환
이쁘게_날짜로변환 = 날짜로변환.strftime("%Y:%m:%d")
이쁘게_날짜로변환


def 날짜변환함수(날짜넣어라):
    for 숫자데이터 in 날짜넣어라:
        if 숫자데이터.isdigit():
            # print(숫자데이터)
            변환_숫자데이터 = int(숫자데이터)

    # 2. 현재 날짜에서 정수형으로 변환한 거 빼기
    날짜로변환 = 현재날짜 - timedelta(days=변환_숫자데이터)

    # 3. 뺀 값을 보기 좋게 변환
    이쁘게_날짜로변환 = 날짜로변환.strftime("%Y:%m:%d")
    return 이쁘게_날짜로변환

날짜변환함수(테스트)
```

### 무한스크롤 해서 전체 댓글 확인할 수 있게 만들기

```python
from selenium import webdriver

driver = webdriver.Chrome()

driver.get("https://www.youtube.com/watch?v=ucDJkDtWcww")
```

```python
# 무한스크롤 알고리즘
    # 현재 윈도우 창 높이만큼 스크롤 해줘.
    # 1번을 반복해줘.
    # 언제까지? 0이 될 때까지 

# 절대 위치 이동 | 딱 1000px만 이동
# driver.execute_script("window.scrollTo(0,1000)")

# 상대 위치 이동 | 현재 위치에서 1000px 이동
# driver.execute_script("window.scrollBy(0, 1000)")

# 전체 스크롤 높이 
    # 유튜브는 SPA라서 DOM 접근이 다르다
    # 다른 곳은 "return document.body.scrollHeight" 입력하면 된다
    # 근데 유튜브는 SPA라서 아래처럼 입력해야 한다.
    #     SPA 페이지인지 알아내는 방법
    #         해시 또는 쿼리 파라미터만 바뀌면 SPA
    #         기본 동영상 URL: https://www.youtube.com/watch?v=abc123xyz
    #         같은 동영상의 특정 시간대로 이동할 때: https://www.youtube.com/watch?v=abc123xyz&t=2m30s
현재스크롤전체높이 = driver.execute_script("return document.documentElement.scrollHeight")


import time
import random
랜덤타임 = random.uniform(1,2)

time.sleep(랜덤타임)

# 상대 위치 이동하기 | 현재스크롤높이만큼
driver.execute_script(f"window.scrollBy(0, {현재스크롤전체높이})")

# 페이지 로딩
time.sleep(랜덤타임)


# '새스크롤높이(전체스크롤높이)' 또 확인
새스크롤높이 = driver.execute_script("return document.documentElement.scrollHeight")
# print(새스크롤높이)


# 조건문으로 처음 전체 스크롤 높이와 나중 전체 스크롤 높이를 비교
    # 0이면 break 해서 while 반복문 멈추게 하기
스크롤높이확인 = 0

# 만약 처음 스크롤 전체 높이가 스크롤높이확인과 일치한다면,
#     브레이크

if 새스크롤높이 == 스크롤높이확인:
    break
```

### 무한스크롤 통합코드

```python
# 무한스크롤 로직
# 최초로 현재 스크롤 전체 높이를 확인한다
# 현재 스크롤 전체 높이만큼 이동한다
# 좀 쉰다
# 현재 스크롤 전체 높이를 확인한다
# 만약 최초의 현재 스크롤 전체 높이가 중간에 확인한 전체 스크롤 전체 높이와 같은지 확인한다
#     같으면 멈추게한다.
#     같지 않으면, 새롭게 업데이트 한 현재 스크롤 전체 높이를 처음 스크롤 전체 높이에 값을 저장한다
#     저장하는 이유는 초기값이 스크롤 이동의 '가이드'가 되기 때문이다.
#     스크롤 높이는 아무리 길어도 계속하다보면 0에 가까워지기 때문이다. 0에 가까워지면 멈추게 하려고 이렇게 하는 거다.

# 그리고 '이동 - 쉬기 - 저장'을 반복한다. 
# 저장한 값은 가이드가 되어야 하기 때문에, while 반복문 밖에 쓴다.

import time
import random
from selenium.webdriver.support.ui import WebDriverWait

현재스크롤전체높이 = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    랜덤타임 = random.uniform(1,2)

    # 상대 위치 이동하기 | 현재스크롤높이만큼
    driver.execute_script(f"window.scrollTo(0, {현재스크롤전체높이})")

    # 페이지 로딩
    time.sleep(랜덤타임)

    # 명시적대기 실시
    WebDriverWait(driver, 10).until(
        lambda driver: driver.execute_script("return document.readyState") == "complete"
    )

    # '새스크롤높이(전체스크롤높이)' 또 확인
    새스크롤높이 = driver.execute_script("return document.documentElement.scrollHeight")

    if 새스크롤높이 == 현재스크롤전체높이:
        break

    현재스크롤전체높이 = 새스크롤높이

time.sleep(랜덤타임)
```

### 0번째 행에 LLM 이용해서 감정분석 해보기

```python
# 데이터프레임 확인
# 0번째 행만 감정분석 해보기


df.head()


# 댓글열에 있는 0번째행
    # loc는 레이블로 행과 열을 다루는 메서드
    # iloc는 숫자로 행과 열을 다루는 메서드
comment = df.loc[0, '댓글']

# content에서 줄바꿈 할 때 오류가 안 나려면 """""" 트리플 쿼터 안에서 내용을 입력한다.
response = client.chat.completions.create(
        model = "gpt-4o-mini-2024-07-18",
        messages = [
            {'role':'system', 'content':"You are a data analyst with 10 years of experience at Google."}, 
            {'role':'user', 'content':f"""This data contains YouTube comments from the video 'President Yoon's impeachment vote fails due to insufficient quorum'. As a data analyst with 10 years of experience at Google, you need to perform sentiment analysis on {comment}. Your response will go in the 'sentiment_analysis' column, with the following response types:
                                        1. If the comment content appears to support impeachment, respond with '탄핵찬성'
                                        2. If the comment content appears to oppose impeachment, respond with '탄핵반대'
                                        3. If the comment content seems unrelated to impeachment, respond with '기타'
                                        Please respond only with '탄핵찬성', '탄핵반대', or '기타'."""} 
        ]  
)

# print(response.choices[0].message.content)

# '감정분석' 열에 0번째 행에 답변을 넣어주세요.
df.at[0, '감정분석'] = response.choices[0].message.content

df.head()
```

### 모든 행에 감정분석 진행하기

```python
# 댓글 컬럼에 있는 모든 행
comment = df.loc[:, '댓글']

def 답변생성기(comment):
    response = client.chat.completions.create(
        model = "gpt-4o-mini-2024-07-18",
        messages = [
            {'role':'system', 'content':"You are a data analyst with 10 years of experience at Google."}, 
            {'role':'user', 'content':f"""This data contains YouTube comments from the video 'President Yoon's impeachment vote fails due to insufficient quorum'. As a data analyst with 10 years of experience at Google, you need to perform sentiment analysis on {comment}. Your response will go in the 'sentiment_analysis' column, with the following response types:
                                        1. If the comment content appears to support impeachment, respond with '탄핵찬성'
                                        2. If the comment content appears to oppose impeachment, respond with '탄핵반대'
                                        3. If the comment content seems unrelated to impeachment, respond with '기타'
                                        Please respond only with '탄핵찬성', '탄핵반대', or '기타'."""} 
            ]  
    )

    return response.choices[0].message.content


# 댓글 열에 있는 곳을 ai가 바라보게 한 후, 답변을 감정분석 열 내에 각 행에 담기

# apply() 안에서는 함수 이름만 작성해도 된다. 함수 이름만 작성해도 각 행에 모든 함수 내용이 전달된다.
# 함수 정의시 아규먼트를 썼는데도 안 써도 되는 이유?
#     각 행의 내용이 자동으로 매개변수 자리로 가서 괜찮다.
df['감정분석'] = df['댓글'].apply(답변생성기)
```
